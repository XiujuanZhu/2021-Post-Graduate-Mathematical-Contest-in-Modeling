```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y2, test_size=0.3)
```

```python
#GBDT作为基模型的特征选择
new_X = SelectFromModel(GradientBoostingRegressor(), max_features =20).fit_transform(x_data, y2)
new_X.shape[1]
print(new_X)
```

```python
x_gb_train, x_gb_test, y_gb_train, y_gb_test = train_test_split(new_X, y2, test_size=0.3)

rf_gb = RandomForestRegressor(n_estimators=200, random_state=0)
rf_gb.fit(x_gb_train, y_gb_train)
y_gb_pred = rf_gb.predict(x_gb_test)

# 评估回归性能
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_gb_test, y_gb_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_gb_test, y_gb_pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_gb_test, y_gb_pred)))
print('Squared R:', rf_gb.score(x_gb_test,y_gb_test))
```

```python
models=[LinearRegression(),KNeighborsRegressor(),SVR(),Ridge(),Lasso(),MLPRegressor(alpha=20),DecisionTreeRegressor(),ExtraTreeRegressor(),XGBRegressor(),RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor(),BaggingRegressor()]
models_str=['LinearRegression','KNNRegressor','SVR','Ridge','Lasso','MLPRegressor','DecisionTree','ExtraTree','XGBoost','RandomForest','AdaBoost','GradientBoost','Bagging']
score_=[]


for name,model in zip(models_str,models):
    print('-----------------------')
    print('开始训练模型：'+name)
    model=model   #建立模型
    model.fit(x_gb_train, y_gb_train)
    y_pred=model.predict(x_gb_test)
    print('Mean Absolute Error:', metrics.mean_absolute_error(y_gb_test, y_pred))
    print('Mean Squared Error:', metrics.mean_squared_error(y_gb_test, y_pred))
    print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_gb_test, y_pred)))
    print('Squared R:', model.score(x_gb_test,y_gb_test))
```

# 随机森林特征选择和回归预测

```python
rf = RandomForestRegressor(n_estimators=200, random_state=0)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)

# 评估回归性能

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('Squared R:', rf.score(x_test,y_test))
```

```python
rfc = RandomForestRegressor(n_estimators=200, random_state=0)  
rfc = rfc.fit(x_train,y_train) 
importances = rfc.feature_importances_
std = np.std([tree.feature_importances_ for tree in rfc.estimators_], axis=0)
indices = np.argsort(importances)[::-1]# Print the feature ranking
print("Feature ranking:")
for f in range(min(20,x_data.shape[1])):    
    print("%2d) %-*s %f" % (f + 1, 30, x_data.columns[indices[f]], importances[indices[f]]))
```

```python
x1 = x_data[['MDEC-23', 'minsssN','maxHsOH','BCUTc-1l','C1SP2','MLFER_A','maxssO','minHBint5','LipoaffinityIndex','minsOH','minHBint10',
             'minHsOH','VC-5','SHsOH','SsOH','XLogP','MLogP','TopoPSA','SHBint6','MDEO-12']]
X_train, X_test, Y_train, Y_test = train_test_split(x1, y2, test_size=0.3)
```

```python
regressor = RandomForestRegressor(n_estimators=200, random_state=0)
regressor.fit(X_train, Y_train)
Y_pred = regressor.predict(X_test)

# 评估回归性能

print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
print('Squared R:', regressor.score(X_test,Y_test))
```

```python
Test_X = Test_X[['MDEC-23', 'minsssN','maxHsOH','BCUTc-1l','C1SP2','MLFER_A','maxssO','minHBint5','LipoaffinityIndex','minsOH','minHBint10',
             'minHsOH','VC-5','SHsOH','SsOH','XLogP','MLogP','TopoPSA','SHBint6','MDEO-12']]
pred_y = RandomForestRegressor(n_estimators=200, random_state=0).fit(X_train, Y_train).predict(Test_X)
print(pred_y)

test_y.iloc[:,1]=pred_y

test_y.to_csv('./test_y.csv',index=False)
```
